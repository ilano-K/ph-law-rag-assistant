```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service 
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException

import time
import pandas as pd
```

## Opens Chrome

```python
# start chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

## Direct to the link

```python
# url to scrape
url = 'https://elibrary.judiciary.gov.ph/republic_acts'

# opens the url 
driver.get(url)
```

## Scraping the Title and URLs

```python
all_ra_links  = []

wait = WebDriverWait(driver, 3)

att_filter = "a[href*='/thebookshelf/showdocs/']"
button   = 'button.dt-paging-button.next'

page = 1

while True:
    
    # Wait for the links to load 
    wait.until(EC.presence_of_element_located(
        (By.CSS_SELECTOR, att_filter)
    ))
    
    # number of links per page
    num_links = len(driver.find_elements(By.CSS_SELECTOR, att_filter))
    
    print(f"Scraping links for page: {page} with total of number of links: {num_links}")
    # loop through index
    for i in range(num_links):
        l = driver.find_elements(By.CSS_SELECTOR, att_filter)[i]
        all_ra_links.append({
            'title': l.text,
            'url': l.get_attribute("href")
        })
    
    # fetch button safely
    while True:
        try:
            next_btn = driver.find_element(By.CSS_SELECTOR, button)
            classes = next_btn.get_attribute('class')
            break 
        except StaleElementReferenceException:
            time.sleep(0.2) # retry window
    
    # check if the button is disabled (no next page)
    if 'disabled' in classes:
        break
    
    page +=1
    # else proceed
    next_btn.click()
    time.sleep(2)

driver.quit()
print(f"collected {len(all_ra_links)} RA links.")
    
```

## Convert to dataframe

```python
df = pd.DataFrame(all_ra_links)
df
```
