```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df = pd.read_csv("all_ra_links")

```

```python
links = df['url'].to_list()

scraped = []
for link in links:
    url = link 
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'html.parser')
    
    print("====================================")
    print(f"Scraping contents of url: {url} ...")
    #republic act
    act_tag = soup.find('h2')
    act = act_tag.get_text(strip=True) if act_tag else None 
    

    #title 
    title_tag = soup.find('h3')
    title = title_tag.get_text(strip=True) if title_tag else None 
    
    # content
    main_content = soup.find("div", class_="single_content")
    clean_text = main_content.get_text(separator='\n', strip=True) if main_content else None 
    
    if not clean_text:
        print(f"Can't scrape main content for url {clean_text}")
        break

    scraped.append({
        "url": url,
        "act": act,
        "title": title,
        "raw_content": clean_text,
    })



```

```python
df_scraped = pd.DataFrame(scraped)
```

```python
df_scraped.to_csv("republic_act.csv", index=False)
```
